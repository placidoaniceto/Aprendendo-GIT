Estudo guiado de git
====================

**Autor:** Daniel R. Cassar



Este notebook apresenta algumas perguntas para você responder durante a aula de Git da disciplina de Práticas em Ciência de Dados. A ideia é que você responda estas questões com suas palavras enquanto assiste a aula ou logo após para sedimentar seu conhecimento.

Desta forma, este notebook não é um notebook didático, mas sim um guia para seus estudos.



## Perguntas para estudo guiado



1.  O que significa versionamento de código e para que serve isso?

2.  O que é e para que serve o git?

3.  O que é um repositório online de git (como, por exemplo, um repositório no GitHub)? Qual a utilidade de termos um repositório online?

4.  Para que serve o arquivo de `README`? Quais são as boas práticas de um arquivo `README`?

5.  Para que serve o arquivo `LICENSE`?

6.  Para que serve o arquivo `.gitignore`?

7.  O que faz o comando `git clone`?

8.  O que faz o comando `git status`?

9.  O que faz o comando `git add`? Como adicionar apenas um arquivo com o comando `git add`? Como adicionar todos os arquivos com o comando `git add`?

10. O que faz o comando `git commit`? Qual a diferença entre o `git add` e o `git commit`?

11. O que faz o comando `git push`?

12. O que faz o comando `git pull`? Quando devemos utilizá-lo?

13. Para que serve criar um ramo (*branch*) novo? Em qual situação você se vê ramificando seu repositório?

14. O que faz o comando `git checkout`?

15. O que é um fork?

16. O que é um *pull request*?

17. Quero contribuir para um repositório que não sou o dono, como fazer isso utilizando de fork e *pull request*?
